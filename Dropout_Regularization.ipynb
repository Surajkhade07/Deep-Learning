```python
X=df.drop(60,axis=1)
```

```python
Y=df[60]
```

```python
# y having categorical values so we conert it 
Y=pd.get_dummies(Y,drop_first=True).astype(int)
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
import tensorflow
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(208,activation='relu',input_shape=(60,)),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    X,Y,epochs=100,batch_size=8
)
```

```python
# we introduce dropout layer to reduce model acuuracy 
model=keras.Sequential([
    keras.layers.Dense(208,activation='relu',input_shape=(60,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X,Y,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,Y_test)
```
